```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import matplotlib.pyplot as plt

df_diabetes = pd.read_csv('diabetes.csv')

# PRÉ-PROCESSAMENTO
X_db = df_diabetes.drop('Outcome', axis=1)
y_db = df_diabetes['Outcome']

# ALGORITMO C4.5 (Entropy)
# Usamos 'entropy' para representar a lógica do C4.5
model_c45_db = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model_c45_db.fit(X_db, y_db)

# RESULTADOS
plt.figure(figsize=(20,10))
plot_tree(model_c45_db, feature_names=X_db.columns, class_names=['Saudável', 'Diabetes'], filled=True)
plt.title("C4.5 (Entropia) - Diabetes")
plt.show()

print("Base de Conhecimento C4.5 - Diabetes:\n")
print(export_text(model_c45_db, feature_names=list(X_db.columns)))
```
